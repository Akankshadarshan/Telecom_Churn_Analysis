**DATA CLEANING**

```python
import pandas as pd
df = pd.read_csv("Telco-Customer-Churn.csv")
```

```python
df.head()
```

```python
print(type(df))
```

```python
df.info()
```

```python
df.columns
```

```python
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors='coerce')
```

```python
df["TotalCharges"].fillna(df["TotalCharges"].median(), inplace=True)
```

```python
print(df["TotalCharges"].dtype)
```

```python
df.info()
```

```python
print(df["TotalCharges"].isnull().sum()) 
```

```python
print(df["Churn"].isnull().sum()) 
```

```python
df["Churn"] = df["Churn"].map({"Yes":1, "No":0})
```

```python
print(df["Churn"].unique())
```

```python
print(df["Churn"].dtype) 
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python


df.duplicated().sum()
```

```python
for col in ["gender","Partner","Dependents","PhoneService","MultipleLines",
            "InternetService","OnlineSecurity","OnlineBackup","DeviceProtection",
            "TechSupport","StreamingTV","StreamingMovies","Contract",
            "PaperlessBilling","PaymentMethod"]:
    df[col] = df[col].astype("category")
```

```python
df.head(10)
```

**EXPLORATORY DATA ANALYSIS**

**Bivariate Analysis**

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
sns.countplot(x="Churn", data=df)
plt.title("Customer Churn Distribution")
plt.show()
churn_percent = df["Churn"].value_counts(normalize=True) * 100


print("\nPercentages:\n", churn_percent)
```

```python
cross = pd.crosstab(df['SeniorCitizen'], df['Churn'])
print(cross)
cross_percent = cross.div(cross.sum(axis=1), axis=0) * 100
print(cross_percent.round(2))
```

```python
cross_percent.plot(kind="bar", stacked=True, figsize=(6,4))
plt.xticks([0,1], ['Not Senior', 'Senior'], rotation=0)
plt.title("Churn % by Senior Citizen (Stacked)")
plt.ylabel("Percentage")
plt.show()
```

```python
sns.boxplot(x="Churn", y="tenure", data=df)
plt.title("Tenure vs Churn")
plt.show()
df["TenureGroup"] = pd.cut(df["tenure"], bins=[0,20,40,60,70], 
                           labels=["0-20","21-40","41-60","61-70"])

churn_by_tenure = df.groupby("TenureGroup")["Churn"].mean() * 100
print(churn_by_tenure)
```

```python

sns.countplot(x="PhoneService", hue="Churn", data=df)
plt.title("Churn by PhoneService")
plt.show()
churn_by_PhoneService = df.groupby("PhoneService")["Churn"].mean() * 100
print(churn_by_PhoneService)
```

```python
sns.countplot(x="MultipleLines", hue="Churn", data=df)
plt.title("Churn by MultipleLines")
plt.show()
churn_by_MultipleLines = df.groupby("MultipleLines")["Churn"].mean() * 100
print(churn_by_MultipleLines)
```

```python
sns.countplot(x="InternetService", hue="Churn", data=df)
plt.title("Churn by Internet Services")
plt.show()
churn_by_InternetService = df.groupby("InternetService")["Churn"].mean() * 100
print(churn_by_InternetService)
```

```python
sns.countplot(x="PaymentMethod", hue="Churn", data=df)
plt.title("Churn by PaymentsMethod")

plt.xticks(rotation=45)   # rotate 45°
plt.xticks(rotation=90)   # rotate 90° (vertical)
plt.show()
churn_by_PaymentsMethod = df.groupby("PaymentMethod")["Churn"].mean() * 100
print(churn_by_PaymentsMethod)
```

```python
sns.boxplot(x="Churn", y="MonthlyCharges", data=df)
plt.title("Monthly Charges vs Churn")
plt.show()
df['MonthlyCharges_bin'] = pd.cut(
    df['MonthlyCharges'],
    bins=[0, 30, 60, 90, 120],   # adjust as per your data
    labels=['Low (0-30)', 'Medium (30-60)', 'High (60-90)', 'Very High (90+)']
)
churn_rate = df.groupby('MonthlyCharges_bin')['Churn'].value_counts(normalize=True).mul(100).rename('percent').reset_index()
churn_rate[churn_rate['Churn'] == 1]   # only churned %
```

```python
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
corr = df[numeric_cols].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
